```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV

data = pd.read_csv('./menu.csv')
data.head()
```

```python
# Save to SQLite database
conn = sqlite3.connect('Mcdonalds.db')
data.to_sql('MCDONALDS_NUTRITION', conn, if_exists='replace', index=False)

# Read data from SQL
df = pd.read_sql("SELECT * FROM MCDONALDS_NUTRITION", conn)
df.head()

# Summary of data
df.describe(include='all')
```

```python
# Sodium vs Category (Swarmplot)

plt.figure(figsize=(12, 6))
plot = sns.swarmplot(x='Category', y='Sodium', data=df)
plt.xticks(rotation=45)
plt.title("Sodium Content by Category")
plt.tight_layout()
plt.show()
```

```python
# Calories vs Category (Swarmplot)

plt.figure(figsize=(12, 6))
plot = sns.swarmplot(x='Category', y='Calories', data=df)
plt.xticks(rotation=45)
plt.title("Calories by Category")
plt.tight_layout()
plt.show()
```

```python
# Max Sodium Info

print("Sodium Description:")
print(df['Sodium'].describe())
```

```python
max_sodium_index = df['Sodium'].idxmax()
max_sodium_item = df.at[max_sodium_index, 'Item']
print(f"🔺 Item with most Sodium: {max_sodium_item} ({df.at[max_sodium_index, 'Sodium']} mg)")
```

```python
# Jointplot: Protein vs Total Fat
sns.jointplot(x='Protein', y='Total Fat', data=df)
plt.show()
```

```python
# Boxplot of Sugar
sns.set_style("whitegrid")
plt.figure(figsize=(10, 4))
sns.boxplot(x=df["Sugars"])
plt.title("Sugar Distribution")
plt.show()
```
